### [C1] How many items do people typically buy in one order? What does the distribution look like?

```python
items_per_order = order_products.groupby('order_id')['add_to_cart_order'].count()

mean_items = items_per_order.mean()
median_items = items_per_order.median()

print(f"Average number of items per order: {mean_items:.2f}")
print(f"Median number of items per order: {median_items:.2f}")

plt.figure(figsize=(10, 6))
sns.histplot(items_per_order, bins=30, kde=True, color='skyblue')
plt.title('Distribution of Number of Items Per Order', fontsize=16)
plt.xlabel('Number of Items', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
