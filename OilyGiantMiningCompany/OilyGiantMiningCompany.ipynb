````python
#Train and test the model for each region
regions = [region_1, region_2, region_3]

# Function to train and test model
def train_and_test(region):
    features = region[['f0', 'f1', 'f2']]
    target = region['product']
    
    # Split into train/test
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=12345)
    
    # Train Linear Regression model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Predict on test set
    predictions = model.predict(X_test)
    
    # Evaluate model
    mse = mean_squared_error(y_test, predictions)
    rmse = mse ** 0.5
    
    return model, rmse, X_test, y_test, predictions

# Train and test for each region
results = [train_and_test(region) for region in regions]

# Unpack results
models, rmses, X_tests, y_tests, preds = zip(*results)

rmses
````

````python
# Step 4: Function to calculate profit
def calculate_profit2(predictions, target):
    """
    Calculate profit based on top 200 predicted wells.
    """
    predictions_sorted = predictions.sort_values(ascending=False)
    selected_points = target[predictions_sorted.index][:200]
    product = selected_points.sum()
    revenue = product * 4500
    return revenue - 1e8

# Step 5: Bootstrapping with corrected revenue estimation
state = np.random.RandomState(12345)
bootstrap_results = []

for pred, y_test in zip(preds, y_tests):
    region_profits = []
    for _ in range(1000):
        # Sample 500 wells
        indices = state.choice(pred.shape[0], size=500, replace=True)
        sample_preds = pd.Series(pred[indices]).reset_index(drop=True)
        sample_targets = y_test.iloc[indices].reset_index(drop=True)

        # Calculate profit from top 200 selected based on prediction ranking
        profit = calculate_profit2(sample_preds, sample_targets)
        region_profits.append(profit)

    bootstrap_results.append(region_profits)

# Step 5.2: Summary statistics
bootstrap_summary = []

for region_profits in bootstrap_results:
    mean_profit = np.mean(region_profits)
    lower_bound = np.percentile(region_profits, 2.5)
    upper_bound = np.percentile(region_profits, 97.5)
    risk_of_loss = (np.array(region_profits) < 0).mean() * 100  # as percentage
    bootstrap_summary.append((mean_profit, lower_bound, upper_bound, risk_of_loss))


# Define region labels
region_labels = ['Region 1', 'Region 2', 'Region 3']

# Create DataFrame
summary_df = pd.DataFrame(bootstrap_summary, columns=[
    'Mean Profit ($)', 'Lower 95% CI ($)', 'Upper 95% CI ($)', 'Risk of Loss (%)'
])
summary_df['Region'] = region_labels

# Reorder columns
summary_df = summary_df[['Region', 'Mean Profit ($)', 'Lower 95% CI ($)', 'Upper 95% CI ($)', 'Risk of Loss (%)']]

display("Bootstrap Profit Summary",summary_df)
````
